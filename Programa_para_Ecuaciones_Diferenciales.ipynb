```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
from sympy import *
import sympy as sp
#m=68.1
#c=12.5
#g=9.81
t, m, c, g=sp.symbols('t, m, c, g')
v=sp.Function('v')(t)
diffeq=Eq(v.diff(t)+(c/m)*v,g) #Aqui escribimos la ecuaciones,
#despues de la coma se toma como un =; por ejemplo aqui estamos igualando a g
display(diffeq)
dsolve(diffeq, v, hint='default', simplify=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0,100,100)
m=68.1
c=12.5
g=9.81
v1 = (g*m)/(c)*(1-np.exp(-c/m)*t)
plt.plot(t,v1,'--r')
plt.title('Grafica velocidad paracaidista')
plt.xlabel('t', color='#1C2833')
plt.ylabel('h', color='#1C2833')
plt.grid()
```
